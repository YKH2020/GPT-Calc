```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df = pd.read_csv("Iris.csv")
X = df.iloc[:,1:5].values.T  # Transpose to match X's shape in the given code
Y = np.zeros((3,150))
Y[0,0:50] = 1
Y[1,50:100] = 1
Y[2,100:150] = 1

plt.figure(figsize=(15, 5))

# Feature 1 vs Feature 2
plt.subplot(1, 3, 1)
plt.scatter(X[0,:50], X[1,:50], color="red", marker="x", label="Class 1")
plt.scatter(X[0,50:100], X[1,50:100], color="blue", marker="o", label="Class 2")
plt.scatter(X[0,100:], X[1,100:], color="green", marker="+", label="Class 3")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.legend()

# Feature 1 vs Feature 3
plt.subplot(1, 3, 2)
plt.scatter(X[0,:50], X[2,:50], color="red", marker="x", label="Class 1")
plt.scatter(X[0,50:100], X[2,50:100], color="blue", marker="o", label="Class 2")
plt.scatter(X[0,100:], X[2,100:], color="green", marker="+", label="Class 3")
plt.xlabel("Feature 1")
plt.ylabel("Feature 3")
plt.legend()

# Feature 1 vs Feature 4
plt.subplot(1, 3, 3)
plt.scatter(X[0,:50], X[3,:50], color="red", marker="x", label="Class 1")
plt.scatter(X[0,50:100], X[3,50:100], color="blue", marker="o", label="Class 2")
plt.scatter(X[0,100:], X[3,100:], color="green", marker="+", label="Class 3")
plt.xlabel("Feature 1")
plt.ylabel("Feature 4")
plt.legend()

plt.suptitle("Scatter plots of the Iris dataset")
plt.show()

# These plots show the distribution of the Iris dataset in each pair of features. In general, it seems that classes 1 and 2 are more similar to each other than to class 3, which is more spread out. 
# However, it isn't clear how well a linear classifier would perform on this dataset without testing.
```
