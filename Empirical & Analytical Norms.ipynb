```python
import numpy as np
from numpy.linalg import norm

m = 9
n = 6

A = np.random.rand(m, n)

num_samples = 100000

max_norm = 0

# Sampling 100,000 random vectors to find the maximum norm
for i in range(num_samples):
    x = np.random.randn(n)

    # The norm of Ax
    Ax_norm = norm(np.dot(A, x))

    # The norm of x
    x_norm = norm(x)

    # The ratio of the norms
    ratio = Ax_norm / x_norm

    # Updating max_norm
    if ratio > max_norm:
        max_norm = ratio

print("Empirical maximum norm:", max_norm)

# The analytically derived maximum norm
U, S, Vt = np.linalg.svd(A)
anal_max_norm = S[0]
print("Analytic maximum norm:", anal_max_norm)
```
