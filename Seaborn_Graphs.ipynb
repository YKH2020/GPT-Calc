```python
%matplotlib inline
import matplotlib.pyplot as plt
plt.style.use('seaborn-whitegrid')
import numpy as np

fig = plt.figure()
ax = plt.axes()

x = np.linspace(0, 10, 1000)
ax.plot(x, np.sin(x));

y = np.linspace(0, 10, 1000)
ax.plot(y, np.cos(y));

ax.set(xlim=(0, 10), ylim=(-2, 2),
       xlabel='x', ylabel='sin(x) & cos(x)',
       title='A Simple Plot');
```

```python
import random

num_sims = 100_000
event_count = 0
values=[]
values2=[]

for sim in range(num_sims):
    die = random.choice(range(1,7))
    die2 = random.choice(range(1,7))
    
    values += [die]
    values2 += [die2]
    
    if values[sim] == 4 & values2[sim] == 4:
        event_count += 1

print('Rolling a Hard Eight: ',event_count/num_sims)
    
```
